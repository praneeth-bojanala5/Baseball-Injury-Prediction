```python
import os
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
raw_df = pd.read_csv('integrated_data.csv', low_memory=False)
raw_df
```

```python
raw_df.drop(columns=['at_bat_number', 'pitch_name', 'sz_top', 'sz_bot'], inplace=True, axis=0)
```

```python
raw_df
```

```python
raw_df.isna().sum()
```

```python
raw_df.dropna(inplace=True)
```

```python
raw_df.isna().sum()
```

```python
count_frame = pd.DataFrame(raw_df[['player_name', 'game_pk']].value_counts())
singles = count_frame[count_frame['count']==1]
singles.reset_index(inplace=True)
singles[['player_name', 'game_pk']]
```

```python
clean_df = pd.merge(raw_df, singles[['player_name', 'game_pk']], how='outer', indicator=True).query('_merge != "both"').drop('_merge', axis=1).reset_index(drop=True)
clean_df
```

```python
clean_df.columns
```

```python

zone_cat = clean_df['zone'].unique()
zone_counts = clean_df['zone'].value_counts(normalize=True)
zone_counts[9]
```

```python
final_df = pd.DataFrame()
i = 0
for id in clean_df['ID'].unique():
    id_df = clean_df[clean_df['ID'] == id]
    
    predictor = id_df[['pitch_of_game']]
    
    regression_columns = ['release_speed', 'release_pos_z',  'release_spin_rate']
    
    to_add_df = id_df[['player_name', 'pitcher', 'game_date', 'game_pk', 'ID', 'game_precedes_injury']].drop_duplicates()
    
    for column in regression_columns:
        response = id_df[[f'{column}']]
        
        regr = linear_model.LinearRegression()
        regr.fit(predictor, response)
        
        slope = regr.coef_[0][0]
        mean = id_df[[f'{column}']].mean(axis=1)
        std = id_df[[f'{column}']].std()[f'{column}']
        
        
        to_add_df[f'{column}_slope'] = slope
        to_add_df[f'{column}_mean'] = mean
        to_add_df[f'{column}_std'] = std
        
        
    # categorical_columns = ['pitch_type', 'zone', 'type']
    # 
    # pitch_type_counts = id_df['pitch_type'].value_counts(normalize=True)
    # 
    # type_counts = id_df['type'].value_counts(normalize=True)
    # 
    # pitch_type_cat = id_df['pitch_type'].unique()
    # zone_cat = id_df['zone'].unique()
    # type_cat = id_df['type'].unique()
    # 
    # for cat in pitch_type_cat:
    #     to_add_df[f'{cat}_proportion'] = pitch_type_counts[cat]
    # 
    # for cat in zone_cat:
    #     to_add_df[f'{cat}_proportion'] = zone_counts[cat]
    # 
    # for cat in type_cat:
    #     to_add_df[f'{cat}_proportion'] = type_counts[cat]
        
    
    final_df = pd.concat([final_df, to_add_df])
        
final_df
```

```python
final_df.to_csv('numerical_columns.csv', index=False)
```

```python
predictor = final_df[['release_speed_slope', 'release_speed_mean', 'release_speed_std', 'release_pos_z_slope', 'release_pos_z_mean', 'release_pos_z_std', 'release_spin_rate_slope', 'release_spin_rate_mean', 'release_spin_rate_std']]
response = final_df['game_precedes_injury']

X_train, X_test, y_train, y_test = train_test_split(predictor, response, test_size=0.20, random_state=42)

log_regr = linear_model.LogisticRegression()

log_regr.fit(X_train, y_train)

y_pred = log_regr.predict(X_test)
y_train_pred = log_regr.predict(X_train)

print(confusion_matrix(y_train, y_train_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)
y_train_pred = gnb.predict(X_train)

print(confusion_matrix(y_train, y_train_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
svc = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)
y_train_pred = svc.predict(X_train)

print(confusion_matrix(y_train, y_train_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
